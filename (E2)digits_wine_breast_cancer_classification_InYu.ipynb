### 1. 데이터 준비
scikit-learn에서 제공하는 toy dataset 중, 분류 문제에 적합한 digits과 wine, breast_cancer을 import하자.

```python
# import 필요한 모듈
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# (1) 데이터 준비
from sklearn.datasets import load_digits
from sklearn.datasets import load_wine
from sklearn.datasets import load_breast_cancer

digits = load_digits()
digits_data = digits.data # Feature Data 지정하기
digits_label = digits.target # Label Data 지정하기

print('digits labels : ', digits.target_names) # Target Names 출력해 보기
print('(data 개수, features)=',digits_data.shape) # 데이터 Describe 해 보기
print('-----------------------------------------------')

wine = load_wine()
wine_data = wine.data # Feature Data 지정하기
wine_label = wine.target # Label Data 지정하기
print('wine labels : ', wine.target_names) # Target Names 출력해 보기
print('(data 개수, features)=',wine_data.shape) # 데이터 Describe 해 보기
print('-----------------------------------------------')

breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data # Feature Data 지정하기
breast_cancer_label = breast_cancer.target # Label Data 지정하기
print('breast_cancer labels : ', breast_cancer.target_names) # Target Names 출력해 보기
print('(data 개수, features)=',breast_cancer_data.shape) # 데이터 Describe 해 보기
```

### 2. 데이터 톺아보기

```python
# (2)-1 digits 데이터 톺아보기
print(digits.DESCR)
```

#### digits.DESCR 설명 (Description of DESCR)
digits 에서 64는 8X8픽셀 값을 일렬로 펴진 데이터를 뜻합니다.  
펴진 data를 reshape하면 handwritten digits 이미지를 확인할 수 있고, 이 데이터가 총 1797개 입니다.  
NIST가 제공하는 사전 처리 프로그램은 미리 인쇄된 양식에서 handwritten digits의 표준화된 비트맵을 추출하는 데 사용됐습니다.  
총 43명 중에서 30명이 훈련 세트에 기여했으며 13명이 시험 세트에 기여했습니다.  
32x32 비트맵은 4x4의 분할되지 않는 블록으로 나뉘고 각 블록에서 픽셀 수가 계산됩니다.(32/4=16)  
이렇게 하면 각 요소가 0~16 범위의 정수인 8x8의 입력 행렬이 생성됩니다.  
이 과정은 차원성을 줄이고 작은 왜곡에 불변성을 부여합니다.

와인의 세 가지 유형에서 발견되는 다른 성분들에 대해 13가지 다른 측정치가 있습니다.

### 3. 문제와 정답준비
pandas 라이브러리를 이용해 2차원 배열로 문제와 정답을 나열합니다.  
문제 : 머신러닝 모델에게 입력되는 데이터. feature라고 부르기도 한다. 변수 이름으로는 X를 많이 사용한다.  
정답 : 머신러닝 모델이 맞추어야 하는 데이터. label 또는 target이라고 부르기도 한다. 변수 이름으로는 y를 많이 사용한다.

```python
# (3) 문제와 정답준비

import pandas as pd

# 준비한 데이터셋을 pandas가 제공하는 DataFrame 이라는 자료형으로 변환 => 문제
# digits
digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names)

# 정답인 label 컬럼 추가(가장 오른쪽에 추가됨)
digits_df["label"] = digits.target
digits_df
```

```python
# wine
wine_df = pd.DataFrame(data=wine_data, columns=wine.feature_names)

# 정답인 label 컬럼 추가(가장 오른쪽에 추가됨)
wine_df["label"] = wine.target
wine_df
```

```python
# breast_cancer
breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names)

# 정답인 label 컬럼 추가(가장 오른쪽에 추가됨)
breast_cancer_df["label"] = breast_cancer.target
breast_cancer_df
```

```python
# 중략된 표에 1인 label이 하나 밖에 보이지 않지만 중간을 살펴보면 뒤섞여있음을 볼 수 있다.
breast_cancer_df[200:300]
```

### 4. train, test 데이터 분리
scikit-learn이 제공하는 train_test_split 이라는 함수로 데이터를 분리한 후 학습시킵니다.

```python
# (4) train, test 데이터 분리
X_digits_train, X_digits_test, y_digits_train, y_digits_test = train_test_split(digits_data, digits_label, test_size=0.2, random_state=7)
print('digits_______________________________________________')
print('X_digits_train 개수:', len(X_digits_train), ', X_digits_test 개수:', len(X_digits_test))

X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(wine_data, wine_label, test_size=0.2, random_state=7)
print('''
wine___________________________________________''')
print('X_wine_train 개수:', len(X_wine_train), ', X_wine_test 개수:', len(X_wine_test))

X_breast_cancer_train, X_breast_cancer_test, y_breast_cancer_train, y_breast_cancer_test = train_test_split(breast_cancer_data, breast_cancer_label, test_size=0.2, random_state=7)
print('''
breast_cancer_____________________________________________________''')
print('X_breast_cancer_train 개수:', len(X_breast_cancer_train), ', X_breast_cancer_test 개수:', len(X_breast_cancer_test))
```

### 5. 다양한 모델로 학습 & 평가 (Separately 1.digits/ 2. wine/ 3.breast_cancer)  

오차 행렬(confusion matrix)은 original label과 predicted label 관계를 나타낸 행렬이다.  
Classification_report는 오차 행렬을 가지고 모델을 평가할 수 있는 척도인 precision과 recall, f1-score, accuracy 총 4가지를 계산한 것이다.  

```python
# 1.digits
from sklearn.metrics import classification_report
# (5)-1.Decision Tree
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_digits_train, y_digits_train)
y_digits_pred = decision_tree.predict(X_digits_test)
print('Decision Tree_________________________________________')

# Confusion matrix(오차 행렬)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_digits_test, y_digits_pred))

print(classification_report(y_digits_test, y_digits_pred))

# (5)-2.Random Forest 사용해 보기
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_digits_train, y_digits_train)
y_digits_pred = random_forest.predict(X_digits_test)
print('Random Forest_________________________________________')

# Confusion matrix(오차 행렬)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_digits_test, y_digits_pred))

print(classification_report(y_digits_test, y_digits_pred))

# (5)-3.SVM 사용해 보기
from sklearn import svm
svm_model = svm.SVC(random_state=32)
svm_model.fit(X_digits_train, y_digits_train)
y_digits_pred = svm_model.predict(X_digits_test)
print('SVM___________________________________________________')

# Confusion matrix(오차 행렬)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_digits_test, y_digits_pred))

print(classification_report(y_digits_test, y_digits_pred))

# (5)-4.SGD Classifier 사용해 보기
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier(random_state=32)
sgd_model.fit(X_digits_train, y_digits_train)
y_digits_pred = sgd_model.predict(X_digits_test)
print('SGD Classifier________________________________________')

# Confusion matrix(오차 행렬)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_digits_test, y_digits_pred))

print(classification_report(y_digits_test, y_digits_pred))

# (5)-5.Logistic Regression 사용해 보기 
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_digits_train, y_digits_train)
y_digits_pred = logistic_model.predict(X_digits_test)
print('Logistic Regression___________________________________')

# Confusion matrix(오차 행렬)
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_digits_test, y_digits_pred))

print(classification_report(y_digits_test, y_digits_pred))
```

wine 결과 평가의 기준도 digits과 비슷하게 precision(정밀도), recall(재현율), f1-score(재현율과 정밀도의 조화평균), accuracy(정확도) 중 어느 것으로 하던지 간에 RandomForest model 하나가 1.00로 가장 높다.
